```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#正規ホワイトノイズ
def normal_white_noise(sigma :float, length: int) -> list:
    noise_list = [np.random.normal(0,sigma^2) for _ in range(length)]
    return noise_list
```

```python
plt.plot(normal_white_noise(1, 300))
```

```python
#ma1モデル
def ma1(mu :float, sigma :float, theta :float, length: int) -> list:
    y_t_list = []
   #初期(t=0)でepsilon_{t-1}=0
    ep_t_list = [0]
    for i in range(length):
        ep_t_list.append(np.random.normal(0,sigma^2))
        y_t_list.append(mu+ep_t_list[i]+theta*ep_t_list[i-1])
    return y_t_list
```

```python
#いろいろなパラメータで探索

thera_list = [0.1,0.5,0.9]
sigma_list = [1,5,25]

fig, ax = plt.subplots(3, 3, figsize=(20, 12),dpi=200)
for i,theta in enumerate(thera_list):
  for j,sigma in enumerate(sigma_list):
    ax[i, j].plot(ma1(mu=1, sigma=sigma, theta=theta, length=200))
    ax[i, j].set_title(f'sigma:{sigma},theta:{theta}')
```
