```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#ma1モデル
def ar1(y_0 :float, c :float, sigma :float, theta :float, length: int) -> list:
    y_t_list = []
    noise_list = np.random.normal(0,sigma,length)
    for i in range(length):
        if i == 0:
            y_t_list.append(y_0)
        else:
            y_t_list.append(c+theta*y_t_list[i-1]+noise_list[i])
    return y_t_list
```

```python
#いろいろなパラメータで探索

thera_list = [0.1,0.5,0.9,1,1.1]
y_0_list = [1,10,100]

fig, ax = plt.subplots(5, 3, figsize=(20, 20),dpi=200)
for i,theta in enumerate(thera_list):
    for j,y_0 in enumerate(y_0_list):
        ax[i, j].plot(ar1(y_0=y_0, c=2, sigma=1, theta=theta, length=1000))
        ax[i, j].set_title(f'y_0:{y_0},theta:{theta}')
```
